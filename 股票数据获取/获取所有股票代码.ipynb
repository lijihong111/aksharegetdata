```python
#-*- codeing = utf-8 -*-
#获取所有股票代码
import akshare as ak
import pandas as pd
import numpy as np
import os
StockList=ak.stock_info_a_code_name()
for i in range(0,len(StockList)):
    temp=StockList.iloc[i,0]
    if temp[0]=="6":
        temp="sh"+temp
    if temp[0]=="0":
        temp="sz"+temp
    if temp[0]=="3":
        temp="sz"+temp
    StockList.iloc[i,0]=temp
#通过txt形式保存数据，先检测是否存在文件。
#若存在，则在更新股票列表时，先将原有文件删除，再重建文件。
if os.path.exists("./data/stockList.csv"):
           os.remove("./data/stockList.csv")
StockList.to_csv("./data/stockList.csv",index=False,header=True)
#将原有表格删除后，更新原有数据
df=pd.read_csv('./data/stockList.csv')
#查看数据数据及分布
df.describe()
df
```

```python
#-*- codeing = utf-8 -*-
#删掉创业板
StockList=pd.read_csv('./data/stockList.csv')
for i in range(0,len(StockList)):
    temp=StockList.iloc[i,0]
    if temp[:5]=="sz300" or temp[:5]=="sz301" or temp[:5]=="sh688"or temp[:5]=="sh689" or temp[:1]!='s':
        StockList.iloc[i,0]='null'
StockList.drop(StockList.index[StockList['code']=="null"],inplace=True)

StockList=StockList.reset_index(drop=True)
#通过txt形式保存数据，先检测是否存在文件。
#若存在，则在更新股票列表时，先将原有文件删除，再重建文件。
if os.path.exists("./data/stockList_del_cyb.csv"):
           os.remove("./data/stockList_del_cyb.csv")
StockList.to_csv("./data/stockList_del_cyb.csv",index=False,header=True)
#将原有表格删除后，更新原有数据
df=pd.read_csv('./data/stockList_del_cyb.csv')
#查看数据数据及分布
df.describe()
df
```
